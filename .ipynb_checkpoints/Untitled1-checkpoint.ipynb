```python
import numpy as N
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('nuala')
```

```python
filepath = '/gpfs/data/llinares/2016/log_density/more_runs_lcdm_corrected_ics/analysis/power_corrected_output_density/runs/'
```

```python
import powerspec as P
```

```python
nruns = 43
ngrid = 1024
lbox = 256.
```

```python
kbins = N.logspace(N.log10(0.02), N.log10(13), 21)
```

```python


for i in N.arange(nruns):
    for j in N.arange(3):
        f=open('%srun_%i_%02i/density_2d_%i.dat'%(filepath, ngrid, i+1, j+1))
        tmp = N.fromfile(f, dtype=N.float, count = 3*ngrid**2)
        tmp = N.reshape(tmp, [ngrid**2, 3])
        dens = N.reshape(tmp[:,2], [ngrid, ngrid])
        dens = dens/N.mean(dens)
        if ((i==0) & (j==0)):
            k, pk = P.pkn(dens, kbins, boxsize = lbox)
            pkdens = N.zeros([pk.size, 3*nruns], dtype=N.float)
            pklog = N.zeros([pk.size, 3*nruns], dtype=N.float)
        k, pkdens[:,3*i+j] = P.pkn(dens, kbins, boxsize = lbox)
        k, pklog[:,3*i+j] = P.pkn(N.log(dens), kbins, boxsize=lbox)
        
        
        
```

```python
# Compute covariance (correlation) matrices of power spectra
corr_dens = N.corrcoef(pkdens)
corr_log = N.corrcoef(pklog)

cov_dens = N.cov(pkdens)
cov_log = N.cov(pklog)
```

```python
plt.figure()
plt.pcolormesh(N.log10(k), N.log10(k), corr_dens)
plt.xlabel('$\log_{10} k$')
plt.ylabel('$\log_{10} k$')
plt.colorbar()
plt.figure()
plt.pcolormesh(N.log10(k), N.log10(k), corr_log)
plt.colorbar()
plt.xlabel('$\log_{10} k$')
plt.ylabel('$\log_{10} k$')
```

```python
pkdens_mean = N.mean(pkdens,1)
pklog_mean = N.mean(pklog,1)

sn_dens = N.zeros_like(pkdens_mean)
sn_log = N.zeros_like(pklog_mean)
```

```python
for j in N.arange(1, k.size, 1):
    cinv_dens = N.linalg.inv(cov_dens[0:j, 0:j])
    cinv_log = N.linalg.inv(cov_log[0:j, 0:j])
    sn_dens[j] = N.dot(pkdens_mean[0:j], N.dot(cinv_dens, pkdens_mean[0:j]))
    sn_log[j] = N.dot(pklog_mean[0:j], N.dot(cinv_log, pklog_mean[0:j]))
```

```python
plt.loglog(k, N.sqrt(sn_dens))
plt.loglog(k, N.sqrt(sn_log))
plt.xlabel(r'$k_{\rm max}$ [$h$ / Mpc]')
plt.ylabel('Signal-to-noise')
```

```python
mg_filepath = '/gpfs/data/llinares/2016/log_density/eight_runs/first_run_symm_a/analysis/power_corrected_output_density/runs/'
```

```python
models = ['lcdm', 'symm_a', 'symm_b', 'symm_c', 'symm_d', 'fofr6', 'fofr5', 'fofr4']
```

```python
def get_bias(pk1, pk2, numinbin):
    ind=N.where(pk1[:,0]<0.09)[0][-1]
    bias3=N.mean(pk2[0:ind, 1]*numinbin[0:ind])/N.mean(pk1[0:ind, 1]*numinbin[0:ind])
    return bias3
```

```python
pkdens_mg = {}
pklog_mg = {}
for mm in ['lcdm', 'symm_a']:
    f=open('%srun_%s_%i/density_2d.dat'%(mg_filepath, mm, ngrid), 'rb')
    tmp = N.fromfile(f, dtype=N.float, count = 3*ngrid**2)
    tmp = N.reshape(tmp, [ngrid**2, 3])
    dens = N.reshape(tmp[:,2], [ngrid, ngrid])
    dens = dens / N.mean(dens)
    k, pkdens_mg[mm], numinbin = P.pkn(dens, kbins, boxsize=lbox, getnuminbin = True)
    
    bias = get_bias(N.transpose([k, pkdens_mg[mm]]), N.transpose([k, pkdens_mg['lcdm']]), numinbin)
    #print bias
    pkdens_mg[mm] = bias*pkdens_mg[mm]
    k, pklog_mg[mm], numinbin = P.pkn(N.log(dens), kbins, boxsize=lbox, getnuminbin=True)
    
    bias = get_bias(N.transpose([k, pklog_mg[mm]]), N.transpose([k, pklog_mg['lcdm']]), numinbin)
    pklog_mg[mm] = bias*pklog_mg[mm]
    #print bias
```

```python
plt.pcolormesh(N.log(dens))
```

```python
plt.loglog(k, pkdens_mg['symm_a'])
plt.loglog(k, pklog_mg['symm_a'], '--')
plt.loglog(k, pkdens_mean)
plt.loglog(k, pklog_mean, '--')
```

```python
signal_to_noise_dens = {}
signal_to_noise_log = {}
for mm in ['lcdm', 'symm_a']:
    signal_to_noise_dens[mm] = N.zeros_like(k)
    signal_to_noise_log[mm] = N.zeros_like(k)
    for j in N.arange(1, k.size, 1):
        cinv_dens = N.linalg.inv(cov_dens[0:j, 0:j])
        cinv_log = N.linalg.inv(cov_log[0:j, 0:j])
        signal_to_noise_dens[mm][j] = N.dot(pkdens_mg[mm][0:j], N.dot(cinv_dens, pkdens_mg[mm][0:j]))
        signal_to_noise_log[mm][j] = N.dot(pklog_mg[mm][0:j], N.dot(cinv_log, pklog_mg[mm][0:j]))
```

```python
plt.loglog(k, N.sqrt(sn_dens), 'k', label='lcdm all')
plt.loglog(k, N.sqrt(sn_log), 'k--')
for mm in ['lcdm', 'symm_a']:
    l1,=plt.loglog(k, N.sqrt(signal_to_noise_dens[mm]), label=mm)
    plt.loglog(k, N.sqrt(signal_to_noise_log[mm]), '--', color=l1.get_color())
plt.legend(loc=2)
```

```python
ax = plt.subplot(111)

log_ratio_err = pklog_mg['symm_a']/pklog_mg['lcdm']*N.sqrt(N.diag(cov_log))*N.sqrt(1/pklog_mg['symm_a']**2 + 1/pklog_mg['lcdm']**2)
dens_ratio_err = pkdens_mg['symm_a']/pkdens_mg['lcdm']*N.sqrt(N.diag(cov_dens))*N.sqrt(1/pkdens_mg['symm_a']**2 + 1/pkdens_mg['lcdm']**2)
plt.errorbar(k, pklog_mg['symm_a']/pklog_mg['lcdm'], yerr = log_ratio_err)
plt.errorbar(k, pkdens_mg['symm_a']/pkdens_mg['lcdm'], yerr = dens_ratio_err)
ax.set_xscale("log")
plt.ylim([0.8, 1.3])
#ax.set_yscale("log")
```
